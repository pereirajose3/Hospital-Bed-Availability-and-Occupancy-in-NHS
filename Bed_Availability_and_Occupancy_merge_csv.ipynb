```python
import pandas as pd

df_occupied = pd.read_csv("C:\\Users\\e410675\\OneDrive - London South Bank University\\LSBU - Jose's Collaborative Projects\\Norfolk and Waveney Demand Modelling\\New Data\\Bed Availability and Occupancy Data – Overnight 2000-2025\\KH03-Occupied-Overnight-only.csv")

# Show the first 5 rows
df_occupied
```

```python
import pandas as pd

df_organisation_codes = pd.read_csv("C:\\Users\\e410675\\OneDrive - London South Bank University\\LSBU - Jose's Collaborative Projects\\Norfolk and Waveney Demand Modelling\\New Data\\Data and scripts\\Organisation Codes.csv")

# Rename one column
df_organisation_codes = df_organisation_codes.rename(columns={'Code': 'Organisation_Code'})


# Show the first 5 rows
df_organisation_codes

```

# Merge DFs of the available and ocupied beds

```python
df_available_occupied = pd.merge(
    df_available,
    df_occupied,
    on=['Organisation_Code', 'Effective_Snapshot_Date', 'Sector'],
    how='inner',
    suffixes=('_available', '_occupied')
)
df_available_occupied
```

```python
df = pd.merge(
    df_organisation_codes,
    df_available_occupied,
    on=['Organisation_Code'],
    how='right',
    suffixes=('_available', '_occupied')
)
df
```

```python
#Order the columns
df = df[[
    'Effective_Snapshot_Date',
    'Organisation_Code',
    'Organisation Name',
    'Sector',
    'Number_Of_Beds_available',
    'Number_Of_Beds_occupied'
]]


#Rename the columns
df=df.rename(columns={
    'Effective_Snapshot_Date': 'Date',
    'Organisation_Code': 'Org_Code',
    'Organisation Name': 'Org_Name',
    'Sector': 'Sector',
    'Number_Of_Beds_available': 'Number_Of_Beds_available',
    'Number_Of_Beds_occupied': 'Number_Of_Beds_occupied'
})

df
```

```python
# Sort the DataFrame by date
df = df.sort_values(by='Date', ascending=True).reset_index(drop=True)
```

```python
# Show only columns with missing values
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]

print(missing_values)
```

```python
# Save to CSV
df.to_csv('Bed_Availability_and_Occupancy_2000-2025_combined_data_csv_format.csv', index=False)
```
